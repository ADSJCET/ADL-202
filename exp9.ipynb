```python
import matplotlib.pyplot as plt

x = range(0,7+1)
y = []
fx = lambda x: x ** 4 + 5

for i in x:
    y.append(fx(i))
    
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = f (x)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 4*np.pi, 1000)
vin = np.sin(t)
vout= np.abs(vin)

plt.xlabel("Time")
plt.xlabel("Volatge")
plt.plot(t, vin, label = 'Input Voltage')
plt.title("Full Wave Rectifier Input")
plt.show()


plt.xlabel('Time')
plt.ylabel('Voltage')
plt.plot(t, vout, label = 'Output Voltage')
plt.title("Full Wave Rectifier Output")
plt.show()
```
