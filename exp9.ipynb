```python
import matplotlib.pyplot as plt

x = range(0,7+1)
y = []
fx = lambda x: x ** 4 + 5

for i in x:
    y.append(fx(i))
    
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = f (x)")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D', 'E', 'F']
y = [1,2,3,4,5,6]
    
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Normal Graph")

plt.show()
```
